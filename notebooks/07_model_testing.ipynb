```python
import joblib
import numpy as np
```

```python
diabetes_model = joblib.load(
    "../models/diabetes_model.pkl"
)

diabetes_scaler = joblib.load(
    "../models/diabetes_scaler.pkl"
)
```

```python
diabetes_input = np.array([
    [2, 180, 80, 30, 150, 35.0, 0.5, 40]
])
```

```python
diabetes_input_scaled = diabetes_scaler.transform(
    diabetes_input
)
```

```python
prediction = diabetes_model.predict(
    diabetes_input_scaled
)

if prediction[0] == 1:
    print("Diabetic")
else:
    print("Non-Diabetic")
```

```python
heart_model = joblib.load(
    "../models/heart_model.pkl"
)

heart_scaler = joblib.load(
    "../models/heart_scaler.pkl"
)
```
